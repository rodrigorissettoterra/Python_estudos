É um processo de caráter investigativo, onde tentamos compreender várias características, como: os valores presentes nas colunas, os tipos de estrutura de dados, verificar se são dados qualitativos ou quantitativos, se há valores faltantes ou incomuns.

Por isso, nesse momento, perguntas sobre os dados são sempre bem-vindas. Elas irão guiar todo o processo de análise, e, através das ferramentas disponíveis, como o nosso querido Pandas, iremos buscar por respostas.

##Qual o percentual de cada tipo de imóvel na nossa base de dados?

#Tratar valores nulos

Como nossos dados serão usados pelo time de ML para criação de um modelo de precificação de imóveis, é importante tratarmos os dados nulos, uma vez que dados nulos não podem ser utilizados no treinamento de modelos de ML.

```python
df_percentual_tipo_ap.isnull().sum()
```

Entende-se que temos:
- 17 valores nulos para a coluna "Valor",
- 4093 valores nulos para a coluna "Condomínio" e
- 10237 valores nulos para a coluna "IPTU".

Escolhi realizar uma abordagem mais simples para tratar os valores nulos, que é substituor tais valores pela média dos valores da coluna.

```python
df_percentual_tipo_ap = df_percentual_tipo_ap.fillna(0)
```

```python
df_percentual_tipo_ap.isnull().sum()
```

Nenhuma das nossas colunas possui valor nulo.

#Remover registros incosistentes

Alguns registros existentes na base de dados não fazem muito sentido, por exemplo:

-    apartamentos que possuem valor de aluguel igual a 0;
-    apartamentos com o valor do condomínio igual a 0.

Esses registros são inconsistentes, por isso devemos removê-los da nossa base de dados.


Procurando o índice de cada linha que precisa ser removida da Base de dados.

```python
registros_a_remover = df_percentual_tipo_ap.query("Valor == 0 | Condominio == 0").index
registros_a_remover
```

```python
df.drop(registros_a_remover, axis = 0, inplace=True)
```

Verificação da exclusão dos registros

```python
df_percentual_tipo_ap.query("Valor == 0 | Condominio == 0")
```

Recebemos um conjunto de dados vazio.

Vamos trabalhar somente com o tipo "Apartamento".

Portanto, podemos reduzir o tamanho do conjunto de dados, excluindo-se a coluno "Tipo".

```python
df_percentual_tipo_ap.drop('Tipo', axis=1, inplace=True)
df_percentual_tipo_ap
```

#Aplicar filtros

Nesse momento, o time de ML deseja avaliar apenas imóveis em cenários específicos, como:

1. Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200;

2. Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000 e área maior que 70 m².

Com base nisso, precisamos aplicar os filtros necessários para obter essas informações específicas.


##Parte 1

Seleção 1

```python
selecao_1 = df_percentual_tipo_ap['Quartos'] == 1
df_percentual_tipo_ap[selecao_1]
```

Seleção 2

```python
selecao_2 = df_percentual_tipo_ap['Valor'] < 1200
df_percentual_tipo_ap[selecao_2]
```

Unindo ambos os filtros

```python
selecao_final = (selecao_1) & (selecao_2)
df_percentual_tipo_ap[selecao_final]
```

Vamos salvar a seleção realizada

```python
df_1 = df_percentual_tipo_ap[selecao_final]
```

##Parte 2

```python
selecao_parte_2 = (df_percentual_tipo_ap['Quartos'] >= 2) & (df_percentual_tipo_ap['Valor'] < 3000) & (df_percentual_tipo_ap['Area'] > 70)
df_percentual_tipo_ap[selecao_parte_2]
```

Salvando o resultado em uma variável

```python
df_2 = df_percentual_tipo_ap[selecao_parte_2]
```

#Criar colunas numéricas

O time de desenvolvedores do site da empresa solicitou a criação de duas novas colunas numéricas na nossa base de dados. No site dos imóveis, eles gostariam de apresentar informações mais sumarizadas em relação a valores. Para isso, essas colunas serão importantes:

-    valor_por_mes: essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio;
-    valor_por_ano: essa coluna deve conter os gastos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio.

**Observação:** Retomaremos a base de dados completa



Criação de coluna "Valor_por_mes".

```python
dados = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv", sep=";")
dados
```

```python
dados['valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados
```

Criação de coluna "Valor_por_ano".

```python
dados['valor_por_ano'] = dados['valor_por_mes'] * 12 + dados['IPTU']
dados
```

O time de desenvolvimento solicitou a criação de mais duas colunas. No entanto, dessa vez, elas são categóricas:
-  Descrição: essa coluna deve possuir uma sumarização das princiáis informações dos imóveis, que serão apresentadas no site: Tipo de imóvel, bairro, quantidade de quartos e vagas de garagem;
-  Possui_suite: essa coluna deve ser uma coluna que informa apenas se o imóvel possui ou não suítes, sem se importar com a quantidade.

Criação da colunba "Descrição"

```python
dados['Descricao'] = dados['Tipo']+ " em " + dados['Bairro'] + \
                    " com " + dados['Quartos'].astype(str) + " quarto(s) " + \
                    " e " + dados['Vagas'].astype(str) + " vagas de garagem."
dados
```

Criando a coluna "possui_suite"

```python
dados['possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados
```

Salvando o novo conjunto de dados

```python
dados.to_csv('dados_completos_dev.csv', index = False)
```
