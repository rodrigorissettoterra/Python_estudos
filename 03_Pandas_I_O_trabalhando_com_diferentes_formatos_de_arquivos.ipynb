#Arquivo em formato .JSON

#Base de dados baixados diretamente de banco de dados

Biliotecas necessárias

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
pip install --upgrade 'sqlalchemy<2.0'
```

Criação do banco de dados

```python
engine = create_engine('sqlite:///:memory')
```

Escrever dados no banco de dados

```python
url = '/content/drive/MyDrive/Base de dados/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
dados
```

```python
dados.to_sql('clientes', engine, index=False)
```

Inspecionar o banco de dados

```python
inspector = inspect(engine)
```

Verificando se conseguimos criar a tabela necessária "clientes"

```python
print(inspector.get_table_names())
```

Consultas

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

##Salvando a base de dados no SQL

```python
empregados.to_sql('empregado', con=engine, index=False)
```

Lendo uma tabela

```python
pd.read_sql('empregado', engine)
```

Leitura de apenas algumas colunas

```python
pd.read_sql('empregado', engine, columns=['ID_Cliente', 'Grau_escolaridade',	'Rendimento_anual'])
```
